```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = (10, 7)
plt.rcParams['figure.facecolor'] = '#FFE5B4'
```

```python
data = pd.read_csv('C:/Users/User/Desktop/Data Science/Data_analysis/world-happiness-report-2021.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
data_columns = ['Country name', 'Regional indicator','Ladder score','Logged GDP per capita', 'Social support', 'Healthy life expectancy','Freedom to make life choices',
    'Generosity', 'Perceptions of corruption']
```

```python
data = data[data_columns].copy()
```

```python
df = data.rename({'Country name':'Country_Name', 'Regional indicator':'Regional_Indicator','Ladder score':'Happiness_Score','Logged GDP per capita':'Logged_GDP_PER_CAPITA', 'Social support':'Social_Support', 'Healthy life expectancy':'Health_Life_Expectancy','Freedom to make life choices':'Freedom_to_make_life_choices',
    'Generosity':'Generosity', 'Perceptions of corruption':'Perception_Of_Corruption'}, axis=1)
```

```python
df.head()
```

# DATA VISUALISATION

# PLOT BETWEEN HAPPINESS AND GDP

```python
df.columns
```

```python
plt.rcParams['figure.figsize'] = (15, 7)
plt.title('Plot between Happiness Score and GDP')
sns.scatterplot(x = df.Happiness_Score, y = df.Logged_GDP_PER_CAPITA, hue = df.Regional_Indicator, s = 200),

plt.legend(loc = 'upper left', fontsize = '10')
plt.xlabel('Happiness Score')
plt.ylabel('GPD PER CAPITA')
```

```python
df.columns
```

## SUM OF ALL THE REGIONS IN TERMS OF GDP PER CAPITA

```python
gdp_region = df.groupby('Regional_Indicator')['Logged_GDP_PER_CAPITA'].sum()
```

```python
gdp_region
```

```python
gdp_region.plot.pie(autopct = '%1.1f%%')
plt.title('GDP BY Region')
plt.ylabel('')
```

```python
df.columns
```

## TOTAL COUNTRIES

```python
total_country = df.groupby('Regional_Indicator')[['Country_Name']].count()
total_country
```

# CO-RELATION MAP

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
cor = numeric_df.corr(method='pearson')

# Create a heatmap
f, ax = plt.subplots(figsize=(10, 5))
sns.heatmap(cor, mask=np.zeros_like(cor, dtype=bool), 
            cmap='Reds', square=True, ax=ax)

plt.show()
```

```python
df.columns
```

```python
corruption = df.groupby('Regional_Indicator')[['Perception_Of_Corruption']].mean()
corruption
```

```python
plt.rcParams['figure.figsize'] = (12, 8)
plt.title('Perception of Corruption in Various Regions')
plt.xlabel('Regions', fontsize = 15)
plt.ylabel('Corruption Index', fontsize = 15)
plt.xticks(rotation = 30, ha = 'right')
plt.bar(corruption.index, corruption.Perception_Of_Corruption)
```

```python
df.columns
```

# TOP 10 LIFE EXPECTANCY AND BOTTOM 10 LIFE EXPECTANCY

```python
top_ten = df.head()
bottom_ten = df.tail()
```

```python
fig, axes = plt.subplots(1,2, figsize= (16, 6))
plt.tight_layout(pad= 2)
xlabels= top_ten.Country_Name
axes[0].set_title('Top 10 Happiest Countries Life Expectancy')
axes[0].set_xticklabels(xlabels, rotation = 45, ha='right')
sns.barplot(x=top_ten.Country_Name, y=top_ten.Health_Life_Expectancy, ax= axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life Expectancy')

xlabels= bottom_ten.Country_Name
axes[1].set_title('Bottom 10 Happiest Countries Life Expectancy')
axes[1].set_xticklabels(xlabels, rotation = 45, ha='right')
sns.barplot(x=bottom_ten.Country_Name, y=bottom_ten.Health_Life_Expectancy, ax= axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('Life Expectancy')

```

## LEVELS OF CORRUPTION IN COUNTRIES

```python
country = df.sort_values(by='Perception_Of_Corruption').head(10)
plt.rcParams['figure.figsize']=(12, 6)
plt.title('Countries with Most Perception of Corruption')
plt.xlabel('Country', fontsize = 13)
plt.ylabel('Corruption Index', fontsize = 13)
plt.xticks(rotation  = 30, ha = 'right')
plt.bar(country.Country_Name, country.Perception_Of_Corruption)
```

# BOTTOM COUNTRIES IN TERMS OF PERDEPTION TO CORRUPTION

```python
country = df.sort_values(by='Perception_Of_Corruption').tail(10)
plt.rcParams['figure.figsize']=(12, 6)
plt.title('Countries with Least Perception of Corruption')
plt.xlabel('Country', fontsize = 13)
plt.ylabel('Corruption Index', fontsize = 13)
plt.xticks(rotation  = 30, ha = 'right')
plt.bar(country.Country_Name, country.Perception_Of_Corruption)
```
