```python
# Function to create bar plots for categorical variables
def plot_bar(variable):
    plt.figure(figsize=(8, 5))
    sns.countplot(x=variable, data=data, palette='Set2')
    plt.title(f'Distribution of {variable}')
    plt.xticks(rotation=45)
    plt.show()

# List of categorical columns to visualiz 
categorical_columns =   ['procurement_system_usage','IT_integration_Supplier','Website_Purchase','Internet_Reliability', 
                       'E_Proccurement_Efficiency', 'Reduces_Costs', 
                       'approvals_workflow_after_eprocurement_adoption', 
                       'E_Procurement_Training', 'compatibility_with_the_existing_systems',
                       'Functionality_Eprocurement', 'Readily_Available', 'Work_Enhancement', 'Need_For_More_Training']

# Generate bar plots for all categorical columns
for col in categorical_columns:
    plot_bar(col)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to create and save bar plots for categorical variables
def plot_bar(variable):
    plt.figure(figsize=(8, 5))
    sns.countplot(x=variable, data=data, palette='Set2')
    plt.title(f'Distribution of {variable}')
    plt.xticks(rotation=45)
    
    # Save the plot as a JPG image
    filename = f"{variable}_distribution.jpg"
    plt.savefig(filename, format='jpg', dpi=300, bbox_inches='tight')
    
    plt.show()

# List of categorical columns to visualize
categorical_columns = [
    'IT_integration_Supplier','Need_For_More_Training','Business_Ease'
]

# Generate and save bar plots for all categorical columns
for col in categorical_columns:
    plot_bar(col)
```

```python
from scipy.stats import chi2_contingency

# Chi-square test on two categorical columns (Procurement Usage vs. E_Proccurement_Efficiency)
contingency_table = pd.crosstab(data['procurement_system_usage'], data['E_Proccurement_Efficiency'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi2 Statistic: {chi2}, p-value: {p}")
```
