1. Business Context Example (Sales Data Visualization)

Letâ€™s assume you have sales data that includes monthly sales for different product categories.

# Python Code (Seaborn and Matplotlib)

```python
# Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data for business context (monthly sales for different product categories)
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Electronics': [15000, 18000, 17000, 20000, 22000, 19000, 23000, 21000, 24000, 25000, 26000, 27000],
    'Clothing': [12000, 11000, 13000, 14000, 15000, 14500, 16000, 15000, 17000, 16500, 17500, 18000],
    'Furniture': [10000, 9000, 11000, 12000, 13000, 12500, 14000, 13000, 14500, 14000, 15000, 15500]
}

# Creating DataFrame
df_sales = pd.DataFrame(data)

# Plotting the sales data
plt.figure(figsize=(10, 6))

# Line plot for sales trends
sns.lineplot(x='Month', y='value', hue='variable', 
             data=pd.melt(df_sales, ['Month']), palette='Set1', linewidth=2.5)

# Customizing the plot
plt.title('Monthly Sales Trends by Product Category', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales Amount (in USD)', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)

# Displaying the plot
plt.tight_layout()
plt.show()
```

Explanation:
Business Context: This example visualizes the monthly sales trends for three different product categories (Electronics, Clothing, and Furniture). The line plot helps analyze how sales change over time and compare performance across categories.
Visualization Tool: Seaborn (for smooth and aesthetically pleasing plots) and Matplotlib (for additional customization).


### 2. Environmental Context Example (Temperature and CO2 Emissions)

Here we visualize the relationship between average global temperatures and CO2 emissions over the years, which is common in environmental studies.

### Python Code (Seaborn and Matplotlib)

```python
# Importing necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data for environmental context (Temperature vs CO2 Emissions)
data_env = {
    'Year': [2000, 2005, 2010, 2015, 2020],
    'Avg_Global_Temperature_C': [14.3, 14.5, 14.6, 14.8, 15.1],
    'CO2_Emissions_GTons': [25.0, 28.5, 32.1, 35.0, 36.5]
}

# Creating DataFrame
df_env = pd.DataFrame(data_env)

# Create a figure with two axes for line and scatter plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting temperature trend (line plot)
sns.lineplot(x='Year', y='Avg_Global_Temperature_C', data=df_env, ax=ax1, color='orange', marker="o")
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Avg Global Temperature (Â°C)', fontsize=12, color='orange')
ax1.tick_params(axis='y', labelcolor='orange')

# Creating a secondary axis to plot CO2 emissions
ax2 = ax1.twinx() 
sns.barplot(x='Year', y='CO2_Emissions_GTons', data=df_env, ax=ax2, alpha=0.6, color='blue')
ax2.set_ylabel('CO2 Emissions (Gigatons)', fontsize=12, color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Adding title and customizing the plot
plt.title('Global Temperature and CO2 Emissions (2000-2020)', fontsize=16)

# Displaying the plot
plt.tight_layout()
plt.show()
```

Explanation:
Environmental Context: The line plot shows the trend in average global temperature over time, while the bar plot on a secondary axis illustrates CO2 emissions in gigatons for the same years. This visual helps in correlating rising temperatures with increasing CO2 emissions, a key environmental concern.
Visualization Tool: Seaborn for both line plot and bar plot, with Matplotlib handling the dual axes to plot two variables on different scales.


### Business Context: Correlation Heatmap (Sales Data)

```python
# Sample sales data for correlation
data_corr = {
    'Electronics': [15000, 18000, 17000, 20000, 22000],
    'Clothing': [12000, 11000, 13000, 14000, 15000],
    'Furniture': [10000, 9000, 11000, 12000, 13000]
}

df_corr = pd.DataFrame(data_corr)

# Plotting a correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df_corr.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title('Correlation between Product Sales', fontsize=14)
plt.show()
```

### Environmental Context: Histogram of Environmental Data

```python
# Sample environmental data (e.g., Rainfall amounts)
import numpy as np

rainfall = np.random.normal(100, 20, 1000)  # Simulated rainfall data

# Plotting histogram of rainfall data
plt.figure(figsize=(8, 5))
sns.histplot(rainfall, bins=30, kde=True, color='green')
plt.title('Distribution of Rainfall Amounts', fontsize=16)
plt.xlabel('Rainfall (mm)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

Explanation:
The heatmap helps visualize correlations between sales categories (business context), and the histogram shows the distribution of rainfall data (environmental context).
