```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame
Forestry = {
    'Achievements': ['First_Quarter', 'Second_Quarter', 'Third_Quarter'],
    'Tree_Planting': [27371, 72285, 1230],
    'Seedling_Production': [13640, 7000, 11250],
    'Awareness' : [3, 4, 5],
    'Capacity_Building_workshop' : [17, 11, 7],
    'Regulatory_Activities' : [316, 267, 424],
    'Agricultural_Shows' : [2, 3, 1]
}

df = pd.DataFrame(Forestry)


# Set the index to Achievements to use it as labels
df.set_index('Achievements', inplace=True)

# Create a bar plot for each achievement category
df.plot(kind='bar', figsize=(10, 6))
plt.title('Forestry Achievements Comparison')
plt.ylabel('Count')
plt.xlabel('Achievements')
plt.xticks(rotation=0)
plt.legend(title="Categories", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
plt.tight_layout()

plt.savefig('Forestry Achievements.png',bbox_inches='tight')
```
