## MACHINE LEARNING

```python
X = pd.get_dummies(df.drop('Condition',axis=1),drop_first=True)
y = df['Condition']
```

### DATA TRAINING

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### DECISION TREE MODEL

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_pred = model.predict(X_test)
```

```python
base_pred #Default hyperparameters
```

### EVALUATION

```python
from sklearn.metrics import confusion_matrix, classification_report,ConfusionMatrixDisplay
```

```python
confusion_matrix(y_test,base_pred) 
```

```python
cm = confusion_matrix(y_test,base_pred)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
```

```python
print(classification_report(y_test,base_pred))
```

```python
model.feature_importances_
```

```python
feature_importance = pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Feature Importance'])
```

```python
feature_importance.sort_values('Feature Importance',ascending=False).iloc[:5]
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8))
plot_tree(model);
```

```python
plt.figure(figsize=(12,8),dpi=150)
plot_tree(model,filled=True,feature_names=X.columns);
plt.title('Decision Tree Model Showing Key Predictors of Borehole Condition')
# Save the graph as an image
plt.savefig('Decision_Tree', dpi=300, bbox_inches='tight')
plt.show()
```

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test,model_preds))
    print('\n')
    plt.figure(figsize=(12,8),dpi=150)
    plot_tree(model,filled=True,feature_names=X.columns);
```

```python
pruned_tree = DecisionTreeClassifier(max_depth=6)
pruned_tree.fit(X_train,y_train)
report_model(pruned_tree)
plt.title('Decision Tree Model Showing Key Predictors of Borehole Condition')
# Save the graph as an image
plt.savefig('Decision_Tree', dpi=300, bbox_inches='tight')
plt.show()
```

# GRID SEARCHING

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Define the hyperparameters to tune
param_grid = {
    'criterion': ['gini', 'entropy'],  # Split quality measures
    'max_depth': [3, 5, 10, None],    # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum samples required to split an internal node
    'min_samples_leaf': [1, 2, 5]     # Minimum samples required at a leaf node
}
```

```python
# Create a GridSearchCV object
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
```

```python
# Fit the grid search to the training data
grid_search.fit(X_train, y_train)
```

```python
# Get the best hyperparameters
print("Best Parameters:", grid_search.best_params_)
```

```python
# Train a Decision Tree with the best parameters
best_clf = grid_search.best_estimator_
```

```python
# Test the model on the test set
y_pred = best_clf.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
```
