# Exercise: Customer Spending and Tips Analysis

Task 1: Load the dataset and explore it.

```python
# Import the required libraries
import seaborn as sns
import pandas as pd
```

```python
# Load the dataset
tips = sns.load_dataset('tips')
```

```python
# View the first few rows of the dataset
tips.head()
```

```python
# Encode categorical variable 'smoker'
tips['smoker'] = tips['smoker'].map({'Yes': 1, 'No': 0})
```

```python
# Encode all categorical variables using pd.get_dummies()
tips_encoded = pd.get_dummies(tips, drop_first=True)
```

```python
# Check for any missing values
tips.isnull().sum()
```

Task 2: Visualize the distribution of total bills.

Objective: Understand the overall distribution of the total bill amounts.

```python
import matplotlib.pyplot as plt
```

```python
# Set Seaborn style
sns.set(style="whitegrid")
```

```python
# Create a histogram of the total bill
plt.figure(figsize=(8, 6))
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title('Distribution of Total Bills', fontsize=16)
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

# Task 3: Analyze the relationship between total bill and tip.

### Objective: Investigate how the tip correlates with the total bill.

```python
# Scatter plot with a regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip Amount', fontsize=16)
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.show()
```

## Task 4: Compare average tips by day of the week.

#### Objective: Determine whether there's any trend in tipping behavior across different days.

```python
# Bar plot of average tip by day
plt.figure(figsize=(8, 6))
sns.barplot(x='day', y='tip', data=tips, estimator=lambda x: sum(x) / len(x), ci=None)
plt.title('Average Tip by Day of the Week', fontsize=16)
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip ($)')
plt.show()
```

## Task 5: Analyze the influence of smoking status on total bills.

#### Objective: See if there’s a difference in spending between smoking and non-smoking customers.

```python
# Boxplot of total bill by smoking status
plt.figure(figsize=(8, 6))
sns.boxplot(x='smoker', y='total_bill', data=tips)
plt.title('Total Bill by Smoking Status', fontsize=16)
plt.xlabel('Smoker')
plt.ylabel('Total Bill ($)')
plt.show()
```

# Task 6: Visualize the relationship between the total bill, tip, and size of the party.

#### Objective: Examine how the number of people in the party influences both the bill and the tip.

```python
# Scatter plot with size representing the number of people in the party
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', size='size', hue='size', data=tips, sizes=(20, 200))
plt.title('Total Bill vs Tip, with Party Size', fontsize=16)
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.legend(title='Party Size', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

# Task 7: Heatmap to show correlation between numeric variables.

#### Objective: Visualize the correlation between variables to understand relationships better.

```python
# Task 7: Correlation matrix
corr = tips_encoded.corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```

# ONE CODE FOR EVERYTHING

```python
# Import the required libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset('tips')

# Encode all categorical variables into dummy variables
tips_encoded = pd.get_dummies(tips, drop_first=True)

# View the first few rows of the encoded DataFrame
print(tips_encoded.head())

# Set Seaborn style
sns.set(style="whitegrid")

# Task 1: Create a histogram of the total bill
plt.figure(figsize=(8, 6))
sns.histplot(tips_encoded['total_bill'], bins=20, kde=True)
plt.title('Distribution of Total Bills', fontsize=16)
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()

# Task 2: Scatter plot with a regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='total_bill', y='tip', data=tips_encoded)
plt.title('Total Bill vs Tip Amount', fontsize=16)
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Amount ($)')
plt.show()

# Task 7: Correlation matrix
corr = tips_encoded.corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
